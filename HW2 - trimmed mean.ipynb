```python
#VARIABLES
i = 0
n = 1
datas=0
listD = []
sumD=0
repeat=0

#PROCESS TO INTRODUCE DATA
for k in range(0,3):
    d = float(input(f"Introduce data{n}:"))                
    listD.append(d)
    sumD += d
    n+=1
    datas+=1
while i < 1:
    print("\n Write 0 to stop process")
    if d != 0:
        d = float(input(f"Introduce data{n}:"))
        listD.append(d)
        sumD += d
        datas+=1
        n+=1
    
    else: i+=1

#CALCULUS
orderedlist=sorted(listD)
print(f"List of Datas:\n {listD}")
print(f"Ordered list :\n {orderedlist}")
a=int(input ("Enter the percentage to apply:%"))
oldPromedio = sumD / datas
print(f"Old Mean:\n {oldPromedio}")
pc=a*.01
pcData=pc*datas
pcAplicated=pcData//1
residue=pcData%1

if residue  > .5:
    total=int(pcAplicated+1)

elif residue <.5:
    total=int(pcAplicated)

else:
    total=int(pcAplicated)
print (f"Number of data eliminated:\n{total}")
datas-=1

for d in range (total):
    orderedlist.pop(0)
    datas-=1

for t in range (total):
    orderedlist.pop(datas)
    datas-=1
datas+=1

#OUT
print (f"New list of datas:\n{orderedlist}")
print (f"Number of datas in the new list:\n{datas}")
NewSumD=sum (orderedlist)
trimmedMean=NewSumD/datas
print (f"Trimmed mean:\n{trimmedMean}")
```
